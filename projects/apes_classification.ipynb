```python
from fastai.vision.all import *
import fastai
from pathlib import Path
from fastcore.all import *
```

```python
path = Path("datasets/apes")
#path.ls()
dls = ImageDataLoaders.from_folder(
    path=path,
    valid_pct=0.2,
    seed=42,
    item_tfms=Resize(150),
    batch_tfms=aug_transforms(mult=1.0),
    batch_tfms_scaled_up=aug_transforms(mult=0.5),
    batch_tfms_scaled_down=aug_transforms(mult=2.0)
)
```

```python
dls.show_batch(max_n=6)
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
#learn.lr_find()
# learn.fine_tune(5)
```

```python
learn.fine_tune(5, base_lr=0.00069)
```

```python
label, _, probs = learn.predict("/Users/siddharthnarolia/Projects/Github/DLPractise/projects/datasets/apes/test/Bonobo (Ana Neema) 4062 - Amanda Carberry Columbus Zoo and Aquarium.webp")
```
