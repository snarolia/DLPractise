```python
from fastai.vision.all import *
```

```python
img = PILImage.create("datasets/dogs/dog.6.jpg")
```

```python
img
```

```python
learn_inf = load_learner("models/cat_classify.pkl")
```

```python
learn_inf.dls.vocab

labels = {False:'Not a Cat', True:'Cat'}

pred_class, _, probability = learn_inf.predict(img)

for l,p in zip(learn_inf.dls.vocab, probability):
    print(f"{labels[l]} : {p: 4f}")
```
